```python
import pandas as pd
import numpy as np
```

```python
students = pd.read_csv('ABHAY UBHALE - students.csv', delimiter='\t')
mean = {"DM":students["DM"].mean(),
       "DSA":students["DSA"].mean(),
       "COA":students["COA"].mean(),
       "PL1":students["PL1"].mean()}

range_max = {"DM":students["DM"].std(),
       "DSA":students["DSA"].std(),
       "COA":students["COA"].std(),
       "PL1":students["PL1"].std()}

range_min = {"DM":students["DM"].std()*0.4,
       "DSA":students["DSA"].std()*0.4,
       "COA":students["COA"].std()*0.4,
       "PL1":students["PL1"].std()*0.4}
```

```python
def normalizeData(data):
    grades = list((((data - 0.4*np.min(data)) / (np.max(data) - 0.4*np.min(data)))*10))
    output = list(map(int, grades))
    return output


pointer_DSA = normalizeData(students["DSA"])
pointer_COA = normalizeData(students["COA"])
pointer_DM = normalizeData(students["DM"])
pointer_PL1 = normalizeData(students["PL1"])

print(pointer_DSA)
print(pointer_DM)
print(pointer_COA)
print(pointer_PL1)
```

```python
creditsDF = pd.read_csv('ABHAY UBHALE - credit.txt', delimiter="\t", header = None)

credits = {"DM":creditsDF[1][0],
       "DSA":creditsDF[1][1],
       "COA":creditsDF[1][2],
       "PL1":creditsDF[1][3]}
print(creditsDF)
print(credits)

SPI = []
for i in range(len(pointer_DM)):
    spi = ((pointer_DSA[i]*credits["DSA"]) +(pointer_DM[i]*credits["DM"]) +(pointer_COA[i]*credits["COA"]) +(pointer_PL1[i]*credits["PL1"]))/(credits["DSA"]+credits["DM"]+credits["COA"]+credits["COA"])
        
    SPI.append(round(spi, 2))
SPI
```

```python
m = ["FF", "FF","FF", "FF", "DD", "CD", "CC", "BC", "BB", "AB", "AA"]

def convertGrade(grade_list):
    for i in range(len(grade_list)):
        grade_list[i] = m[(grade_list[i])]


convertGrade(pointer_DSA)
convertGrade(pointer_COA)
convertGrade(pointer_DM)
convertGrade(pointer_PL1)

print(pointer_DSA)
print(pointer_DM)
print(pointer_COA)
print(pointer_PL1)
```

```python
df = {"RollNo":students["RollNo"],
      "Name":students["Name"],
      "DSA":pointer_DSA,
      "DM":pointer_DM,
      "COA":pointer_COA,
      "PL1":pointer_PL1,
      "SPI":SPI}
data = pd.DataFrame(df)
data
```

```python
data.to_csv("results.csv")
```
